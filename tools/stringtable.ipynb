```python
from glob import glob
import xml.etree.ElementTree as ET
import sys
import subprocess
```

```python
#for file in glob('/cygdrive/e/Program Files/games/Steam/steamapps/common/Arma 3/*/*/*/*/stringtable.xml'):
#    print(file)
for file in glob('../addons/*/stringtable.xml'):
    print(file)
    root = ET.parse(file).getroot()
    for package in root:
        for entry in package:
            print(entry.attrib['ID'], entry.find('English').text)
            proc = subprocess.Popen('./stringtable.sh external find "' + entry.find('English').text + '"', shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            for line in proc.stdout:
                print(line.decode('UTF-8','ignore'), sep='')
            print()
```
